## Clustering and evaluation

## K-Means

```python
from sklearn.cluster import KMeans
kmeans_clusters = 5 #(3 is a good upper bound from pca pair plots and going till 5)
fig, ax = plt.subplots(1,kmeans_clusters, figsize=(15,4))
for i in range(2,kmeans_clusters+2): 
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pca_data)
    ax[i-2].scatter(pca_data[:,0], pca_data[:,1], c=kmeans.labels_)
```

```python
from sklearn.metrics import silhouette_score

clusters = [2,3,4,5,6,7,8,9]
n_rep = 10

kmeans_sil = [[0]*len(clusters)]*n_rep
kmeans_cal = [[0]*len(clusters)]*n_rep
kmeans_davis = [[0]*len(clusters)]*n_rep

gmm_sil = [[0]*len(clusters)]*n_rep
gmm_cal = [[0]*len(clusters)]*n_rep
gmm_davis = [[0]*len(clusters)]*n_rep

#Kmeans
for n in range(0,n_rep):
    for i in range(0,len(clusters)):
        kmeans = KMeans(n_clusters=clusters[i])
        kmeans.fit(pca_data)
        kmeans.labels_    
        kmeans_sil[n][i] = silhouette_score(pca_data,kmeans.labels_ )
        kmeans_cal[n][i] = calinski_harabasz_score(pca_data,kmeans.labels_)
        kmeans_davis[n][i] = davies_bouldin_score(pca_data,kmeans.labels_)

#GMM
for n in range(0,n_rep):
    for i in range(0,len(clusters)):
        gmm = mixture.GaussianMixture(n_components=clusters[i])
        gmm.fit(pca_data)
        gmm_predictions = gmm.predict(pca_data)
        gmm_sil[n][i] = silhouette_score(pca_data,gmm_predictions)
        gmm_cal[n][i] = calinski_harabasz_score(pca_data,gmm_predictions)
        gmm_davis[n][i] = davies_bouldin_score(pca_data,gmm_predictions)
```

```python
def calculateMean(vector):
    meanVector = np.mean(vector,axis=0)
    return meanVector
```

```python
#calculate mean of all scores over n_rep iterations

mean_KMeansSil = calculateMean(kmeans_sil)
mean_gmmSil = calculateMean(gmm_sil)
mean_KMeansCal = calculateMean(kmeans_cal)
mean_gmmCal = calculateMean(gmm_cal)
mean_KMeansDavis = calculateMean(kmeans_davis)
mean_gmmDavis = calculateMean(gmm_davis)
```

```python
plt.plot(clusters,mean_KMeansCal,label="KMeans")
plt.plot(clusters,mean_gmmCal,label="GMM")
plt.title('calinski harabasz score plot')
plt.xlabel("clusters")
plt.ylabel("calinski harabasz score")
plt.legend()
```

```python
plt.plot(clusters,mean_KMeansSil,label="KMeans")
plt.plot(clusters,mean_gmmSil,label="GMM")
plt.title('Sillhoute score plot')
plt.xlabel("clusters")
plt.ylabel("Silhouette  Score")
plt.legend()
```

```python
plt.plot(clusters,mean_KMeansDavis,label="KMeans")
plt.plot(clusters,mean_gmmDavis,label="GMM")
plt.title('davies bouldin score plot')
plt.xlabel("clusters")
plt.ylabel("davies bouldin score")
plt.legend()
```

```python
C_All = []
flattened_C = []
```

```python
for k in range(2,11):
    M = np.zeros((len(pca_data),len(pca_data)))
    J = np.zeros((len(pca_data),len(pca_data)))

    #M = np.zeros(len(new_mat))
    for b in range(20):
        row_index = random.sample(range(len(pca_data)), math.floor(0.8*len(pca_data)))
        new_mat = pca_data[row_index[:],:]
        kmeans_task5 = KMeans(n_clusters = k)
        labels = kmeans_task5.fit(new_mat).predict(pca_data)

        for i in row_index:
            for j in row_index:
                #if i in row_index and j in row_index:
                if(labels[i]==labels[j]):
                    M[i][j]+=1
                J[i][j]+=1 

    #print(M)
    #print(J)
    C = M/J
    #print(C)
    C_All.append(C)
    x = C.flatten()
    x.sort()
    flattened_C.append(x)
C_All
```

```python
import seaborn as sns
def ecdf(data):
    """ Compute ECDF """
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)
plt.figure(figsize=(15,10)) 

PAC_scores = np.zeros(len(flattened_C))

for d in range(len(flattened_C)):
    x,y = ecdf(flattened_C[d])
    plt.plot(x,y)
    PAC_df = pd.DataFrame(x,y)
    PAC_scores[d] = y[math.floor(0.99 *len(PAC_df))]-y[math.floor(0.01*len(PAC_df))]
#plt.figure(figsize=(12,8))    
#f.set_figwidth(10)
#f.set_figheight(4)
plt.legend(['K=2','K=3','K=4','K=5','K=6','K=7','K=8','K=9','K=10'])
plt.show()
```
